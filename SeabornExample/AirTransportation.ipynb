```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def scatter_plot(data):
    """"
     Р”РёР°РіСЂР°РјРјР° СЂР°СЃСЃРµСЏРЅРёСЏ
     """
    data.head()
    sns.scatterplot(data=data, color='r', x="year", y="passengers")
    plt.grid(True)
    plt.show()
```

```python
flights_data = sns.load_dataset("flights")  # Р—Р°РіСЂСѓР·РєР° РґР°С‚Р°СЃРµС‚Р° РёР· Seaborn
scatter_plot(flights_data)
```

==========================================================================

```python
def line_plot(data):
    """ Р›РёРЅРµР№РЅС‹Р№ РіСЂР°С„РёРє
    """
    sns.lineplot(data=data, x="year", y="passengers")
    plt.grid(True)
    plt.show()
```

```python
line_plot(flights_data)
```
