```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_heatmap(data):
    """
    РўРµРїР»РѕРІР°СЏ РєР°СЂС‚Р° РјР°С‚СЂРёС†С‹ 5 РЅР° 8
    """
    plt.figure(figsize=(8, 6))
    sns.heatmap(data, vmin=0, vmax=1, annot=True, cmap='coolwarm')
    plt.title('РўРµРїР»РѕРІР°СЏ РєР°СЂС‚Р°')
    plt.show()
```

```python
data = np.random.rand(5, 8)
plot_heatmap(data)
```

```python
def plot_corr_heatmap(data1):
    """
    РўРµРїР»РѕРІР°СЏ РєР°СЂС‚Р° РєРѕСЂСЂРµР»СЏС†РёРѕРЅРЅРѕР№ РјР°С‚СЂРёС†С‹ 6 РЅР° 6
    """
    df = pd.DataFrame(data1)
    correlation_matrix = df.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('РўРµРїР»РѕРІР°СЏ РєР°СЂС‚Р° РєРѕСЂСЂРµР»СЏС†РёРѕРЅРЅРѕР№ РјР°С‚СЂРёС†С‹')
    plt.show()
```

```python
data1 = np.random.rand(6, 6)
plot_corr_heatmap(data1)
```
