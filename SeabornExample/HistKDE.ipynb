```python
def plot_hist_kde(data):
    plt.figure(figsize=(10, 6))
    sns.histplot(data, kde=True, color='b', bins=30)
    plt.title('Р“РёСЃС‚РѕРіСЂР°РјРјР° Рё KDE')
    plt.xlabel('Р—РЅР°С‡РµРЅРёРµ')
    plt.ylabel('Р§Р°СЃС‚РѕС‚Р°')
    plt.grid(True)
    plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)
plot_hist_kde(data)
```
