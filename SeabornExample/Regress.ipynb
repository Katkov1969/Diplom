```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_scatter_with_regression(x, y):
    sns.set(style="darkgrid")  # РЈСЃС‚Р°РЅРѕРІРєР° СЃС‚РёР»СЏ Seaborn
    plt.figure(figsize=(10, 6))
    sns.regplot(x=x, y=y, color='m')
    plt.title('Р”РёР°РіСЂР°РјРјР° СЂР°СЃСЃРµСЏРЅРёСЏ СЃ СЂРµРіСЂРµСЃСЃРёРѕРЅРЅРѕР№ Р»РёРЅРёРµР№')
    plt.xlabel('Р—РЅР°С‡РµРЅРёРµ X')
    plt.ylabel('Р—РЅР°С‡РµРЅРёРµ Y')
    plt.grid(True)
    plt.show()
```

Р’С‹Р·РѕРІ

```python
x = np.random.rand(100)
y = x + np.random.rand(100)
```

```python
plot_scatter_with_regression(x, y)
```
