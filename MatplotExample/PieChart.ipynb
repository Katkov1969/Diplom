```python
import matplotlib.pyplot as plt
```

```python
def plot_pie_chart(labels, sizes):
    plt.figure(figsize=(7, 7))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title('РљСЂСѓРіРѕРІР°СЏ РґРёР°РіСЂР°РјРјР°')
    plt.axis('equal')  # РЈСЂР°РІРЅРёРІР°РµРј РѕСЃРё РґР»СЏ РєСЂСѓРіРѕРІРѕР№ С„РѕСЂРјС‹
    plt.show()
```

Р’С‹Р·РѕРІ С„СѓРЅРєС†РёРё

```python
labels = ['РљР°С‚РµРіРѕСЂРёСЏ A', 'РљР°С‚РµРіРѕСЂРёСЏ B', 'РљР°С‚РµРіРѕСЂРёСЏ C', 'РљР°С‚РµРіРѕСЂРёСЏ D']  # CРїРёСЃРѕРє РєР°С‚РµРіРѕСЂРёР№
sizes = [15, 30, 45, 10]  # РЎРїРёСЃРѕРє Р·РЅР°С‡РµРЅРёР№
plot_pie_chart(labels, sizes)
```
