```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_line_chart(x_values, y_values):
    """
     Р›РёРЅРµР№РЅС‹Р№ РіСЂР°С„РёРє РЅР° РѕСЃРЅРѕРІРµ РїСЂРµРґРѕСЃС‚Р°РІР»РµРЅРЅС‹С… РґР°РЅРЅС‹С… x Рё y.
     """
    plt.figure(figsize=(10, 6))  # РЈСЃС‚Р°РЅР°РІР»РёРІР°РµРј СЂР°Р·РјРµСЂ РіСЂР°С„РёРєР°
    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='Р”Р°РЅРЅС‹Рµ_1')  # РЎС‚СЂРѕРёРј РіСЂР°С„РёРє
    plt.title('Р›РёРЅРµР№РЅС‹Р№ РіСЂР°С„РёРє')  # Р—Р°РіРѕР»РѕРІРѕРє РіСЂР°С„РёРєР°
    plt.xlabel('РћСЃСЊ X')  # РњРµС‚РєР° РѕСЃРё X
    plt.ylabel('РћСЃСЊ Y')  # РњРµС‚РєР° РѕСЃРё Y
    plt.grid(True)  # Р’РєР»СЋС‡Р°РµРј СЃРµС‚РєСѓ
    plt.legend()  # Р”РѕР±Р°РІР»СЏРµРј Р»РµРіРµРЅРґСѓ
    plt.show()  # РћС‚РѕР±СЂР°Р¶Р°РµРј РіСЂР°С„РёРє
```

Р’С‹Р·РѕРІ С„СѓРЅРєС†РёРё

```python
x = [1, 2, 3, 4, 5]
y = [4, 6, 8, 14, 18]
```

Р’С‹Р·РѕРІ С„СѓРЅРєС†РёРё

```python
plot_line_chart(x, y)
```

```python
def plot_two_chart(x_value, y_value, k_value, z_value):
    """
    Р”РІР° РіСЂР°С„РёРєР° РЅР° РѕРґРЅРѕР№ РїР»РѕСЃРєРѕСЃС‚Рё
    """
    plt.style.use('ggplot')
    plt.figure(figsize=(10, 6))  # РЈСЃС‚Р°РЅР°РІР»РёРІР°РµРј СЂР°Р·РјРµСЂ РіСЂР°С„РёРєР°
    plt.plot(x_value, y_value, marker='o', linestyle='-', color='b', label='Р”Р°РЅРЅС‹Рµ_1')  # РЎС‚СЂРѕРёРј РіСЂР°С„РёРє 1
    plt.plot(k_value, z_value, marker='>', linestyle='--', color='r', label='Р”Р°РЅРЅС‹Рµ_2', markerfacecolor='k')  # РіСЂР°С„РёРє 2
    plt.title('Р”РІР° РіСЂР°С„РёРєР° РЅР° РїР»РѕСЃРєРѕСЃС‚Рё ', fontsize=16, fontweight='bold')  # Р—Р°РіРѕР»РѕРІРѕРє РіСЂР°С„РёРєР°
    plt.xlabel('РћСЃСЊ X', fontsize=14, fontweight='bold')  # РњРµС‚РєР° РѕСЃРё X
    plt.ylabel('РћСЃСЊ Y', fontsize=14, fontweight='bold')  # РњРµС‚РєР° РѕСЃРё Y
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=12)
    plt.grid(True)  # Р’РєР»СЋС‡Р°РµРј СЃРµС‚РєСѓ
    plt.legend()  # Р”РѕР±Р°РІР»СЏРµРј Р»РµРіРµРЅРґСѓ
    plt.show()
```

def plot_dual_axis(x_values, y1_values, y2_values):<br>
  
  Р“СЂР°С„РёРє СЃ РґРІСѓРјСЏ РѕСЃСЏРјРё Y


```python
    fig, ax1 = plt.subplots(figsize=(10, 6))
    ax1.set_xlabel('Р’СЂРµРјСЏ')
    ax1.set_ylabel('РўРµРјРїРµСЂР°С‚СѓСЂР°', color='r')
    ax1.plot(x_values, y1_values, color='r')
    ax1.tick_params(axis='y', labelcolor='r')
    ax2 = ax1.twinx()  # РЎРѕР·РґР°РµРј РІС‚РѕСЂСѓСЋ РѕСЃСЊ Y, РєРѕС‚РѕСЂР°СЏ РґРµР»РёС‚ РѕСЃСЊ X СЃ ax1
    ax2.set_ylabel('Р’Р»Р°Р¶РЅРѕСЃС‚СЊ', color='b')
    ax2.plot(x_values, y2_values, color='b')
    ax2.tick_params(axis='y', labelcolor='b')
    fig.tight_layout()  # РЈСЃС‚СЂР°РЅСЏРµРј РЅР°Р»РѕР¶РµРЅРёРµ РјРµС‚РѕРє
    plt.title('Р“СЂР°С„РёРє СЃ РґРІСѓРјСЏ РѕСЃСЏРјРё Y')
    plt.show()
```

Р’С‹Р·РѕРІ С„СѓРЅРєС†РёРё

```python
x = np.arange(0, 10, 1)
y1 = [30, 32, 34, 33, 31, 29, 28, 27, 30, 32]
y2 = [75, 70, 72, 68, 65, 64, 70, 72, 74, 73]
plot_dual_axis(x, y1, y2)
```
