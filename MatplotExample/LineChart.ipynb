```python
import matplotlib.pyplot as plt
import numpy as np

print(plt.style.available) #   

"""
    Matplotlib
"""

"""******************************************************************************************************"""
"""
          x  y.
"""
def plot_line_chart(x_values, y_values):

    plt.figure(figsize=(10, 6))  #   

    plt.plot(x_values, y_values, marker='o', linestyle='-', color='b', label='_1')  #  

    plt.title(' ')  #  
    plt.xlabel(' X')  #   X
    plt.ylabel(' Y')  #   Y
    plt.grid(True)  #  
    plt.legend()  #  
    plt.show()  #  


#  
x = [1, 2, 3, 4, 5]
y = [4, 6, 8, 14, 18]

#  
plot_line_chart(x, y)


''' *************************************************************************************************'''

"""
        
"""

def plot_two_chart(x_value, y_value, k_value, z_value):

    plt.style.use('ggplot')

    plt.figure(figsize=(10, 6))  #   
    plt.plot(x_value, y_value, marker='o', linestyle='-', color='b', label='_1')  #   1
    plt.plot(k_value, z_value, marker='>', linestyle='--', color='r', label='_2', markerfacecolor='k')  #  2

    plt.title('    ', fontsize=16, fontweight='bold')  #  
    plt.xlabel(' X', fontsize=14, fontweight='bold')  #   X
    plt.ylabel(' Y', fontsize=14, fontweight='bold')  #   Y
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=12)
    plt.grid(True)  #  
    plt.legend()  #  

    plt.show()


"""    """
x = [-2, -1, 0, 1, 2, 3, 4, 5, 6, 7]
y = [0.1, 0.3, 0.6, 1.0, 1.4, 1.3, 1.8, 2.0, 1.6, 2.2]

"""    """
k_min = -2
k_max = 10
dk = 0.2

""" C    X     [k_min; K_max],  """
k = np.arange(k_min, k_max + dk, dk)
z = np.sin(0.2 * k) * np.cos(k)

#  
plot_two_chart(x, y, k, z)


''' ********************************************************************************************************'''
"""      Y"""

def plot_dual_axis(x_values, y1_values, y2_values):
    plt.style.use('seaborn-v0_8-colorblind')
    fig, ax1 = plt.subplots(figsize=(10, 6))

    ax1.set_xlabel('')
    ax1.set_ylabel('', color='r')
    ax1.plot(x_values, y1_values, color='r')
    ax1.tick_params(axis='y', labelcolor='r')

    ax2 = ax1.twinx()  #    Y,    X  ax1
    ax2.set_ylabel('', color='b')
    ax2.plot(x_values, y2_values, color='b')
    ax2.tick_params(axis='y', labelcolor='b')

    fig.tight_layout()  #   
    plt.title('    Y', fontsize=14)
    plt.show()


#  
x = np.arange(0, 10, 1)
y1 = [30, 32, 34, 33, 31, 29, 28, 27, 30, 32]
y2 = [75, 70, 72, 68, 65, 64, 70, 72, 74, 73]
plot_dual_axis(x, y1, y2)
```
