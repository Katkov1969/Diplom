def lines_points_3D(x, y, z):<br>
    # РЎРѕР·РґР°РµС‚СЃСЏ С„РёРіСѓСЂР° Рё 3D-РѕСЃРё<br>
    fig = plt.figure(figsize=(8, 6))<br>
    ax = fig.add_subplot(111, projection='3d')<br>
    # РЎС‚СЂРѕСЏС‚СЃСЏ С‚РѕС‡РєРё<br>
    ax.scatter(x, y, z, c='b', marker='o')<br>
    # РўРѕС‡РєРё СЃРѕРµРґРёРЅСЏСЋС‚СЃСЏ Р»РёРЅРёСЏРјРё<br>
    for i in range(len(x)):<br>
        ax.plot([0, x[i]], [0, y[i]], [0, z[i]], color='r', alpha=0.3)<br>
    # РЈСЃС‚Р°РЅР°РІР»РёРІР°СЋС‚СЃСЏ РјРµС‚РєРё Рё Р·Р°РіРѕР»РѕРІРѕРє<br>
    ax.set_xlabel('X')<br>
    ax.set_ylabel('Y')<br>
    ax.set_zlabel('Z')<br>
    ax.set_title('3D РўРѕС‡РєРё Рё Р»РёРЅРёРё')<br>
    plt.show()<br>
# Р“РµРЅРµСЂРёСЂСѓСЋС‚СЃСЏ СЃР»СѓС‡Р°Р№РЅС‹Рµ С‚РѕС‡РєРё<br>

z = np.random.normal(size=100)<br>
# Р’С‹Р·РѕРІС„СѓРЅРєС†РёРё<br>

def plot_3d_surface(x, y, z):<br>
    fig = plt.figure(figsize=(10, 6))<br>
    ax = fig.add_subplot(111, projection='3d')<br>
    ax.plot_surface(x, y, z, cmap='viridis')  # Р¦РІРµС‚РѕРІР°СЏ СЃС…РµРјР°<br>
    ax.set_title('3D РїРѕРІРµСЂС…РЅРѕСЃС‚РЅС‹Р№ РіСЂР°С„РёРє')<br>
    ax.set_xlabel('РћСЃСЊ X')<br>
    ax.set_ylabel('РћСЃСЊ Y')<br>
    ax.set_zlabel('РћСЃСЊ Z')<br>
    plt.show()<br>
# Р’С‹Р·РѕРІ С„СѓРЅРєС†РёРё<br>
x = np.linspace(-5, 5, 100)<br>
y = np.linspace(-5, 5, 100)<br>
x, y = np.meshgrid(x, y)<br>
z = np.cos(np.sqrt(x ** 2 + y ** 2))<br>
plot_3d_surface(x, y, z)<br>

def plot_3d_wireframe(x, y, z):<br>
    fig = plt.figure(figsize=(10, 6))<br>
    ax = fig.add_subplot(111, projection='3d')<br>

    ax.set_xlabel('РћСЃСЊ X')<br>
    ax.set_ylabel('РћСЃСЊ Y')<br>
    ax.set_zlabel('РћСЃСЊ Z')<br>
    plt.show()<br>
# Р’С‹Р·РѕРІ С„СѓРЅРєС†РёРё<br>
