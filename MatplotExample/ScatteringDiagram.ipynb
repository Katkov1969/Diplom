```python
def plot_scatter(x_values, y_values):
    plt.figure(figsize=(10, 6))
    plt.scatter(x_values, y_values, color='r', marker='x')
    plt.title('Р”РёР°РіСЂР°РјРјР° СЂР°СЃСЃРµСЏРЅРёСЏ')
    plt.xlabel('РћСЃСЊ X')
    plt.ylabel('РћСЃСЊ Y')
    plt.grid(True)
    plt.show()
```

```python
x = np.random.rand(1, 100)  # Р“РµРЅРµСЂР°С†РёСЏ СЃР»СѓС‡Р°Р№РЅС‹С… С‡РёСЃРµР»
y = np.random.rand(1, 100)
plot_scatter(x, y)
```
