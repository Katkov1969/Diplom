def plot_multiple_subplots(x, y1, y2):<br>
    fig, axs = plt.subplots(2, 1, figsize=(10, 8))<br>
    axs[0].plot(x, y1, 'r', label='sin(x)')<br>
    axs[0].set_title('РџРѕРґРіСЂР°С„РёРє 1')<br>
    axs[0].set_ylabel('sin x')<br>
    axs[0].legend()<br>
    axs[0].grid(True)<br>
    axs[1].plot(x, y2, 'b', label='cos(x)')<br>
    axs[1].set_title('РџРѕРґРіСЂР°С„РёРє 2')<br>
    axs[1].set_xlabel('РћСЃСЊ X')<br>
    axs[1].set_ylabel('cos x')<br>
    axs[1].legend()<br>
    axs[1].grid(True)<br>
    plt.tight_layout()<br>
    plt.show()<br>
# РІС‹Р·РѕРІ<br>
x_min = -4<br>
x_max = 4<br>
dx = 0.1<br>
x = np.arange(x_min, x_max + dx, dx)<br>
y1 = np.sin(x)<br>
y2 = np.cos(x)<br>
plot_multiple_subplots(x, y1, y2)
