```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
```

```python
def animate_sine_wave():  # С„СѓРЅРєС†РёСЏ Р±СѓРґРµС‚ СЃРѕР·РґР°РІР°С‚СЊ Рё РѕС‚РѕР±СЂР°Р¶Р°С‚СЊ Р°РЅРёРјР°С†РёСЋ СЃРёРЅСѓСЃРѕРёРґР°Р»СЊРЅРѕР№ РІРѕР»РЅС‹
    fig, ax = plt.subplots()  # РЎРѕР·РґР°РµС‚ С„РёРіСѓСЂСѓ (fig) Рё РЅР°Р±РѕСЂ РѕСЃРµР№ (ax).
    x = np.linspace(0, 2 * np.pi, 100)  # РЎРѕР·РґР°РµС‚ РјР°СЃСЃРёРІ x РёР· 100 СЂР°РІРЅРѕРјРµСЂРЅРѕ СЂР°СЃРїСЂРµРґ. С‚РѕС‡РµРє РѕС‚ 0 РґРѕ (2pi).
    """
    РЎРѕР·РґР°РµС‚СЃСЏ РЅР°С‡Р°Р»СЊРЅС‹Р№ РіСЂР°С„РёРє СЃРёРЅСѓСЃРѕРёРґР°Р»СЊРЅРѕР№ РІРѕР»РЅС‹. ax.plot() РІРѕР·РІСЂР°С‰Р°РµС‚ СЃРїРёСЃРѕРє Р»РёРЅРёР№,
    Рё Р·Р°РїСЏС‚Р°СЏ РїРѕСЃР»Рµ line СЂР°СЃРїР°РєРѕРІС‹РІР°РµС‚ РµРіРѕ. np.sin(x) РІС‹С‡РёСЃР»СЏРµС‚ СЃРёРЅСѓСЃ РґР»СЏ РєР°Р¶РґРѕРіРѕ Р·РЅР°С‡РµРЅРёСЏ РІ РјР°СЃСЃРёРІРµ x.
    """
    line, = ax.plot(x, np.sin(x))
    def update(frame):  # С„СѓРЅРєС†РёСЏ Р±СѓРґРµС‚ РІС‹Р·С‹РІР°С‚СЊСЃСЏ РґР»СЏ РѕР±РЅРѕРІР»РµРЅРёСЏ РіСЂР°С„РёРєР° РЅР° РєР°Р¶РґРѕРј С€Р°РіРµ Р°РЅРёРјР°С†РёРё.
        """
        РћР±РЅРѕРІР»СЏРµС‚ РґР°РЅРЅС‹Рµ РїРѕ РѕСЃРё Y Р»РёРЅРёРё, СЃРјРµС‰Р°СЏ СЃРёРЅСѓСЃРѕРёРґР°Р»СЊРЅСѓСЋ РІРѕР»РЅСѓ РЅР° frame / 10.0 СЂР°РґРёР°РЅ.
        Р­С‚Рѕ СЃРѕР·РґР°РµС‚ СЌС„С„РµРєС‚ РґРІРёР¶РµРЅРёСЏ РІРѕР»РЅС‹. Р’РѕР·РІСЂР°С‰Р°РµС‚ РѕР±РЅРѕРІР»РµРЅРЅС‹Р№ РѕР±СЉРµРєС‚ Р»РёРЅРёРё. Р—Р°РїСЏС‚Р°СЏ РЅРµРѕР±С…РѕРґРёРјР° РґР»СЏ
        РІРѕР·РІСЂР°С‚Р° РІ РІРёРґРµ РєРѕСЂС‚РµР¶Р°, С‡С‚Рѕ С‚СЂРµР±СѓРµС‚СЃСЏ РґР»СЏ РїСЂР°РІРёР»СЊРЅРѕР№ СЂР°Р±РѕС‚С‹ СЃ blit
        """
        line.set_ydata(np.sin(x + frame / 10.0))  #
        return line,
    ani = animation.FuncAnimation(fig, update, frames=np.arange(0, 100), blit=True)  # РЎРѕР·РґР°РµС‚ РѕР±СЉРµРєС‚ Р°РЅРёРјР°С†РёРё ani.
    """
    fig: Р¤РёРіСѓСЂР°, РЅР° РєРѕС‚РѕСЂРѕР№ Р±СѓРґРµС‚ РѕС‚РѕР±СЂР°Р¶Р°С‚СЊСЃСЏ Р°РЅРёРјР°С†РёСЏ.
    update: Р¤СѓРЅРєС†РёСЏ, РєРѕС‚РѕСЂР°СЏ Р±СѓРґРµС‚ РІС‹Р·С‹РІР°С‚СЊСЃСЏ РґР»СЏ РѕР±РЅРѕРІР»РµРЅРёСЏ РіСЂР°С„РёРєР°.
    frames=np.arange(0, 100): Р“РµРЅРµСЂРёСЂСѓРµС‚ РїРѕСЃР»РµРґРѕРІР°С‚РµР»СЊРЅРѕСЃС‚СЊ РєР°РґСЂРѕРІ РѕС‚ 0 РґРѕ 99.
    blit=True: РћРїС‚РёРјРёР·РёСЂСѓРµС‚ РїСЂРѕС†РµСЃСЃ Р°РЅРёРјР°С†РёРё, РїРµСЂРµСЂРёСЃРѕРІС‹РІР°СЏ С‚РѕР»СЊРєРѕ С‚Рµ С‡Р°СЃС‚Рё, РєРѕС‚РѕСЂС‹Рµ РёР·РјРµРЅРёР»РёСЃСЊ.
    """
    plt.grid(True)
    plt.show()
```
